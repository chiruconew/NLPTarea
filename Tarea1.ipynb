```python
import pandas as pd
from collections import Counter
import re
import matplotlib.pyplot as plt
```

```python
def rutina_separadoc(inicio, contador, texto):
    doctemp  = ''
    acumula  = ''
    temporal = ''
    for i in range(inicio,len(prueba)):
        temporal = texto[i]
        temporal = temporal.lower()
        acumula  = acumula + temporal
        doctemp  = re.split('\W+',acumula)
        if len(doctemp) > contador:
            break
    
    return(i,doctemp)
```

```python
def rutina_agregadoc(inicio, contador, texto):
    for h in range(0,4):
        resultado = rutina_separadoc(inicio, contador, texto)
        inicio   = resultado[0] + 1
        contador = contador*(5)
        calcula.append(resultado[1])

    return(calcula)
```

```python
prueba = pd.read_csv("C:/Users/chiruco/Desktop/python/ProyPython/tutorial_scraping/prueba.csv")
texto = prueba['body']
calcula  = list()
contador = 10000
inicio   = 0

diccionarios = rutina_agregadoc(inicio, contador, texto)
```

```python
labels = ['Dicc. 1', 'Dicc. 2']
tokens = [len(diccionarios[0]), len(diccionarios[1])]
vocabu = [len(Counter(diccionarios[0])),len(Counter(diccionarios[1]))]

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, tokens, width, label='Tokens')
ax.bar(labels, vocabu, width, label='Vocabulario')

ax.set_ylabel('Vocabulario / Tokens')
ax.set_title('Diccionario 1 y Dicccionario 2')
ax.legend()

plt.show()
```
